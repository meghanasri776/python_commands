Model Building

```python
import statsmodels.formula.api as smf
```

```python
model_1 = smf.ols("DeliveryTime~SortingTime",data = del_time).fit()
```

```python
model_1.params
```

```python
model_1.tvalues , model_1.pvalues
```

```python
model_1.rsquared , model_1.rsquared_adj
```

```python
model_1.summary()
```

```python
model_2 = smf.ols("np.log(DeliveryTime)~SortingTime",data = del_time).fit()
model_2.summary()
```

```python
model_2.rsquared , model_2.rsquared_adj
```

```python
model_3 = smf.ols("np.log(DeliveryTime)~np.log(SortingTime)",data = del_time).fit()
model_3.summary()
```

```python
model_3.rsquared , model_3.rsquared_adj
```

```python
del_time["SortingTime_sq"]= del_time.SortingTime**2
model_4=smf.ols("np.log(DeliveryTime)~np.log(SortingTime_sq+SortingTime)",data=del_time).fit()
model_4.summary()
```

```python
model_4.rsquared , model_4.rsquared_adj
```

Model Prediction

```python
a=int(input('Enter the sorting time:'))
newtime=pd.Series([a])
dt_pred=pd.DataFrame(newtime,columns=["SortingTime"])
b=model_1.predict(dt_pred)
print("the delivery time :",b)
```

```python
delivery_data = pd.read_csv("delivery_time.csv")
print(delivery_data.columns)

X_delivery = delivery_data[['Sorting Time']]
y_delivery = delivery_data['Delivery Time']


X_delivery_train, X_delivery_test, y_delivery_train, y_delivery_test = train_test_split(X_delivery, y_delivery, test_size=0.2, random_state=42)

delivery_model = LinearRegression()
delivery_model.fit(X_delivery_train, y_delivery_train)

y_delivery_pred = delivery_model.predict(X_delivery_test)
rmse_delivery = np.sqrt(mean_squared_error(y_delivery_test, y_delivery_pred))

X_salary = salary_data[['YearsExperience']]
y_salary = salary_data['Salary']

X_salary_train, X_salary_test, y_salary_train, y_salary_test = train_test_split(X_salary, y_salary, test_size=0.2, random_state=42)

salary_model = LinearRegression()
salary_model.fit(X_salary_train, y_salary_train)

y_salary_pred = salary_model.predict(X_salary_test)
rmse_salary = np.sqrt(mean_squared_error(y_salary_test, y_salary_pred))

print(f"RMSE for Delivery_time prediction: {rmse_delivery}")
print(f"RMSE for Salary_hike prediction: {rmse_salary}")
```
