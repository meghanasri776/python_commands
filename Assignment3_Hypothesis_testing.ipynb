Hypothesis Testing - Cutlet Data Exercise

```python
import pandas as pd
import numpy as np
from scipy import stats
from scipy.stats import norm
```

```python
c=pd.read_csv('Cutlets.csv')
c.head()
```

```python
c.shape
```

```python
c.describe()
```

```python
c.isnull().sum
```

```python
unit_A=pd.Series(c.iloc[:,0])
unit_A
```

```python
unit_B=pd.Series(c.iloc[:,1])
unit_B
```

```python
# 2-sample 2-tail t-test:
p_value=stats.ttest_ind(unit_A,unit_B)
p_value
```

```python
#prob
p_value[1]
```

5% is the Significane value and p_value is 47%

Since P_value is greater than significance value.
The Null Hypothesis Ho is True. This shows that there is no difference in the diameter of the Cutlet data between 2 units.

LabTAT data Exercise

```python
tat= pd.read_csv('LabTAT.csv')
```

```python
tat.head()
```

```python
tat.tail()
```

```python
tat.describe()
```

```python
tat.isnull().sum()
```

```python
Lab1=pd.Series(tat.iloc[:,0])
Lab2=pd.Series(tat.iloc[:,1])
Lab3=pd.Series(tat.iloc[:,2])
Lab4=pd.Series(tat.iloc[:,3])
```

```python
pvalue = stats.f_oneway(Lab1,Lab2,Lab3,Lab4)
pvalue[1]
```

```python
pval= 2.1156708949992414e-57
alpha0=0.05
print('significance=%.2f, p=%.2f' %(alpha0, pval))
if pval <= alpha0:
    print('There is a difference in average TAT among the different laboratories & reject the null hyopthesis')
else:
    print('There is no difference in average TAT among the different laboratories & accept the null hypothesis')
```

Buyer Ratio Data Exercise
* Null Hypothesis[H0] : All the proportions are equal.
* Alternative Hypothesis [H1] : Not All the proportions are equal.


```python
from scipy.stats import chi2_contingency
br = pd.read_csv('BuyerRatio.csv')
```

```python
br.head()
```

```python
br.isnull().sum()
```

```python
br.describe()
```

```python
br1=br.set_index('Observed Values')
```

```python
sts, p, dof, expected = chi2_contingency(br1)
p
```

```python
alpha = 0.05
print('significance=%.2f, p=%.2f' %(alpha, p))
if p <= alpha:
    print('Reject Null Hypothesis(Not all proportions are equal)')
else:
    print('Accept Null Hypothesis Ho(All proportions are equal)')
```
